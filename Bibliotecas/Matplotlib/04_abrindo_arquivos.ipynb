# Abrindo arquivos

Uma coisa muito útil do Pyhton é que ele pode abrir arquivos .txt e .csv. Veremos aqui como abrir esses dados e como fazer um gráfico a partir deles.

Lembrando que é preciso já ter o arquivo criado antes, para que ele possa ser aberto e lido.

### Primeira forma 

A primeira forma de abrir um dado é importando a biblioteca csv e trabalhando com ela, como pode ser visto no código abaixo. 

```python
import matplotlib.pyplot as plt
import csv

x = []
y = []

with open("exemplo.txt", "r") as file:
    plots = csv.reader(file, delimiter=",")

    for row in plots:

        x.append(int(row[0]))
        y.append(int(row[1]))
```

O que fizemos aqui foi abrir o arquivo com o comando with e demos um apelido para ele. Lemos o arquivo com o comando csv.reader( ) e em seguida colocamos os valores da coluna 1 do arquivo na lista x e a coluna 2 para a lista y.

Com os dados nas listas, o processo de criação do gráfico é a mesma, podendo fazer o que quiser com eles.

```python
plt.plot(x,y, label="dados vindos do arquivo")

plt.xlabel("x")
plt.ylabel("y")

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.title("Título do gráfico")

plt.legend()

plt.show()
```

### A Segunda forma

A segunda forma é com a biblioteca numpy. É uma maneira bem mais facil de abrir, mas é bom saber mais de uma, por isso mostrei a forma passada. Vamos ver como ela funciona.

```python
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt("exemplo.txt", delimiter=",", unpack=True)
```

Bem mais simples não é? Abrimos o arquivo com o comando np.loadtxt( ), colocamos que o delimitador é a vírgula e desempacotamos eles para as 
variáveis x e y.

O Processo para criação do gráfico é a mesma. 

```python
plt.plot(x,y,label="dados vindos do arquivo")

plt.xlabel("x")
plt.ylabel("y")

plt.title("Título do gráfico")

plt.legend()

plt.show()
```

Perceba que os gráficos são identicos.
