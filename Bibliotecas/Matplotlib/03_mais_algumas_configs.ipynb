# Mais algumas configurações 

#### Mudar estilo do gráfico no decorrer dele

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10, 100)

y = x**2

plt.plot(x[:51], y[:51])
plt.plot(x[50:], y[50:], 'r--')

plt.show()
```

#### Salvar figuras

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,1000)

y = x**3

plt.plot(x, y, color='orange')

plt.title("f(x) = x³")

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.savefig("xaocubo.png", dpi=300) ##

plt.show()
```

### Mudar tamanho da imagem

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,1000)

y = x**(1/2)

plt.figure(figsize = (3,2)) ##

plt.plot(x, y, color='orange')

plt.title("f(x) = x**(1/2)")

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.show()
```
