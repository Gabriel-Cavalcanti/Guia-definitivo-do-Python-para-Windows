# Introdução ao Matplotlib

Segundo o própio site da biblioteca: "Matplotlib é uma biblioteca abrangente para a criação de visualizações estáticas, animadas e interativas em Python. O Matplotlib torna as coisas fáceis fáceis e difíceis possíveis."

#### Instalar o matplotlib

A primeira coisa que é preciso fazer é a instalação da biblioteca. Para isso utilizamos o comando: **pip install matplotlib** ou **conda install matplotlib**, se estiver utilizando o Anaconda.

#### Importar a biblioteca

Uma vez feita a instalação, basta importar a biblioteca para dentro do código. Utilizamos o comando mostrado abaixo. É uma boa prática sempre importar a biblioteca nas primeira linhas do código, sempre! Damos um apelido: "plt" para facilitar a escrita. 

```python
import matplotlib.pyplot as plt
```

#### Plotar o gráfico

Para plotar um gráfico é preciso utilizar dois comando, mostrados abaixo. O primeiro vai criar o gráfico e o outro vai mostrar 
na tela. 

Todo gráfico 2D possui um eixo x e um eixo y, Colocamos o vetor, ou a função, o intervalo, dentro do argumento da 
função plot(x, y), onde x é o eixo x e o y, eixo y.

```python
plt.plot([1,2,4,6],[5,3,6,3])
plt.show()
```

##### Inserindo legenda, titulo e nome dos eixos 

Vamos melhorar um pouco o código e gráfico que acabamos de criar, inserindo um título para ele, o nome dos eixos e uma legenda. Veja o código abaixo.

Obs: O comando para inserir a legenda só ira funcionar caso tenhamos colocado o parâmetro opcional label dentro do plt.plot(x,y, label=).

```python
x = [1,2,4,6]
y = [5,3,6,3]

plt.plot(x,y, label="Isso é a legenda")

plt.title("Esse é o título do gráfico")

plt.xlabel("Esse é o eixo x")
plt.ylabel("Esse é o eixo y")

plt.legend()

plt.show()
```

##### Fazendo algumas configurações no gráfico

```python
x = [1,2,4,6]
y = [5,3,6,3]

plt.plot(x,y, label="Isso é a legenda", color="g", linewidth=3, marker='.', ls='--', markersize=10, markeredgecolor='b')

plt.title("Esse é o título do gráfico")

plt.xlabel("Esse é o eixo x")
plt.ylabel("Esse é o eixo y")

plt.xticks([0, 1, 2, 3, 4, 5, 6])
plt.yticks([3, 4, 5, 6])

plt.legend()

plt.grid(linewidth=5)

plt.show()
```

```python
#  Podemos fazer uma forma mais simples de notação:
# fmt = '[marker][line][color]'
plt.plot(x,y, '*-.b', label="Isso é a legenda")
plt.show()
```

##### Editar o título

Podemos fazer isso com os eixos tmb. Mesmo processo. Coloca o argumento fontdict no xlabel ou ylabel.

```python
x = [1,2,4,6]
y = [5,3,6,3]

plt.plot(x,y, label="Isso é a legenda", color="c", linewidth=2, marker='*', ls='-.', markersize=10, markeredgecolor='g')

plt.title("Esse é o título do gráfico", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xlabel("Esse é o eixo x")
plt.ylabel("Esse é o eixo y")

plt.xticks([0, 1, 2, 3, 4, 5, 6])
plt.yticks([3, 4, 5, 6])

plt.legend()

plt.show()
```
