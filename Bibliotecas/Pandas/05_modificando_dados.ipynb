# Modificando os dados

```python
import pandas as pd

df_csv = pd.read_csv("pokemon_data.csv")

df_csv.head()
```

### Acrecentar uma coluna ao data frame

```python
# Primeira forma
df_csv["Total"] = df_csv['HP'] + df_csv['Attack'] + df_csv['Defense'] + df_csv['Sp. Atk'] + df_csv['Sp. Def'] + df_csv['Speed']

# Segunda forma
df_csv["Total"] = df_csv.iloc[:, 4:10].sum(axis=1)

df_csv.head()
```

### Deletar uma coluna no data frame

```python
df_csv = df_csv.drop(columns=["Total"])

df_csv.head()
```

### Mudar a coluna de posição 

```python
df_csv["Total"] = df_csv.iloc[:, 4:10].sum(axis=1)

df_csv.head()
```

```python
cols = list(df_csv.columns) # pega todas as colunas do data frame e coloca elas numa lista

df_csv = df_csv[cols[0:4] + [cols[-1]] + cols[4:12]]

df_csv.head()
```
