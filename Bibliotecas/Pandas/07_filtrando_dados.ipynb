# Filtrando dados

```python
import pandas as pd

df_csv = pd.read_csv("modified.csv")
```

```python
df_csv.head()
```

Já fizemos algumas espécies de filtro com os dados, utilizando a função .iloc( ). Vamos agora ver mais algumas possibilidades, utilizando a função loc( ).

### Filtro com uma condição

```python
df_water = df_csv.loc[(df_csv["Type 1"] == "Water")]

df_water.head()
```

```python
df_hp_50h = df_csv.loc[(df_csv["HP"] >= 50)]

df_hp_50h.head()
```

### Filtro com mais de uma condição

```python
df_50h_water = df_csv.loc[(df_csv["HP"] >= 50) & (df_csv["Type 1"] == "Water")]

df_50h_water.head()
```

Se quisermos, podemos fazer um sorting nos elementos

```python
df_50h_water.sort_values("HP", ascending=False)
```

### Resetar os indíces

Quando fazemos filtros, o pandas não reseta os índices automaticamente. É preciso fazer isso manualmente, como mostrado abaixo.

```python
df_50h_water.reset_index(drop=1, inplace=True) # Drop: retira índice anterior do data frame, inplace: salva o data frame modificado nele mesmo.

df_50h_water
```

### Maneiras de filtrar mais avançadas

```python
df_nomega = df_csv.loc[~(df_csv["Name"].str.contains("Mega"))]

df_nomega.sort_values("Name")[0:5]
```

```python
import re

df_pi = df_csv.loc[df_csv["Name"].str.contains("^pi[a-z]*", flags=re.I, regex=True)]

df_pi.reset_index(drop=1, inplace=True)

df_pi
```

```python
df_csv.loc[df_csv['Total'] > 500, ['Generation','Legendary']] = ['Test 1', 'Test 2']

df_modificado = df_csv

df_modificado
```
