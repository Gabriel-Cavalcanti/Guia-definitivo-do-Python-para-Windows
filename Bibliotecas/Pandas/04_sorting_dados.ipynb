# Sorting

```python
import pandas as pd

df_csv = pd.read_csv("pokemon_data.csv")
```

Podemos ordenar o nosso data frame da maneira que quisermos. A primeira forma de fazer isso é utilizando o método sort_values( ).

```python
df_csv.sort_values("Name")
```

```python
df_csv.sort_values("Type 1", ascending=False)
```

```python
df_csv.sort_values(["Name", "HP"], ascending=[0,1]).head()
```
