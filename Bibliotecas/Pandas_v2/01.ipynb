### Import the libary

In pandas, there are two types op data: Series and dataframe. The first one is a list with index and a datatype, lika an array (need to be the same datatype). In the other hand, dataframe is a collection od series link by the same index, like a table in a database.

#### Example of a Series

```python
nums = [1, 2, 5, 42, 99, 64, 3]

series1 = pd.Series(nums)
series1
```

Look, there is a index in the left, the data in the right and the datatype in the botton.

#### Exemple of a dataframe

```python
things_qte = [("car", 5), ("bicycle", 4), ("boat", 10), ("motocycles", 2)]

dataframe1 = pd.DataFrame(things_qte)

dataframe1
```

We can name the columns. Instead of 0,1, ... we can put meaning names. To do that, add a parameter in the DataFrame funcion "column=" and add the names inside a list. See bellow

```python
dataframe2 = pd.DataFrame(things_qte, columns=["Vehicles","Quantity"])
dataframe2
```

We can extract the columns for the dataframe, creating a series, doing that

```python
dataframe2['Vehicles']
```

```python
type(dataframe2['Vehicles']) # Notes, it's a Series.
```

We can find the data type of each column in the dataframe. Use the command bellow.

```python
dataframe2.dtypes
```

### Subsetting Data

In this part, we gonna simple select witch data from the dataframe we want. Creating a new dataframe from the preview dataframe.

The first thing is list all the columns from the dataframe.

```python
df.columns
```

The first method is creating a list with all the column that we want and create a "series" with.

```python
df_new = df[['title', 'description', 'publishTime', 'duration_seconds', 'viewCount', 'likeCount', 'commentCount']]
df_new.head()
```

we can see what are the number of columns and rows for any dataframe.

```python
df_new.shape
```

For subsetting rows, we can use two methods: loc or .query

```python
df_subsetrow1 = df_new.loc[df_new['viewCount'] > 1_000_000]
df_subsetrow1
```

```python
df_subsetrow2 = df_new.query('viewCount > 1000000')
df_subsetrow2
```

There is a method that show to us if the value is none or not and returns True od False. We can use that to remove rows with no useful information.

```python
df_new['viewCount'].isna()
```

```python
df_new_nonan = df_new.loc[~df_new['likeCount'].isna()]
df_new_nonan
```

### Casting dtypes

```python
df_new_nonan.dtypes
```

```python
df_new_nonan['viewCount'] = df_new_nonan['viewCount'].astype('int64')
df_new_nonan['likeCount'] = df_new_nonan['likeCount'].astype('int64')
```

```python
df_new_nonan['publishTime'] = pd.to_datetime(df_new_nonan['publishTime'])
df_new_nonan['likeCount'] = pd.to_numeric(df_new_nonan['likeCount'])
```

```python
df_new_nonan.dtypes
```

### Creating a new column

```python
df_new_nonan['like_to_view_ratio'] = (df_new_nonan['likeCount'] /  df_new_nonan['viewCount'] ) * 100
```

```python
df_new_nonan.head()
```

### plot your data

```python
df_new_nonan['viewCount'].plot(kind='hist', bins=50, title='title', figsize=(15,5))
```

```python
df_new_nonan.plot(kind='scatter', x='viewCount', y='likeCount', title='title')
```
